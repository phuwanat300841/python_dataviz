## Select only some colum เลือกคอลัมภ์ที่เราต้องการ ตัดตารางแนวตั้ง

```python
small_table = w_restaurant_df[['name','city_name','zipcode','weighted_average_rating']] 
```

```python
small_table.head()
```

```python
rearch_click_df.head()
```

## Add colum from another table ตัดคอลัมภ์ของอีกตารางมาแปะกับอีกตาราง

```python
small_table['id'] = w_restaurant_df['id'] ## เป็นการเพิ่มคอลัมภ์ที่ชื่อ id โดยเอาค่ามาจากตาราง w_restaurant_df จากฟิลล์ id
```

```python
small_table.head()
```

### .map() เพิ่มคอลัมภ์ในตารางมาอีก 1 คอลัมภ์ 

### เพิ่ม business name ไปในตาราง rearch_click_df

```python
rearch_click_df.head()
```

```python
rearch_click_df['business_name'] = rearch_click_df['business_id'].map(small_table.set_index('id')['name']) 
## การสร้างคอลัมภ์ใหม่ให้กับตาราง ใส่ค่าโดยใช้คำสั่ง .map
##โดย เอาคีย์จากตารางหลัก mapกับตารางรอง โดยเอาค่าจากตารางsmall_table ที่ฟิลล์ name
```

```python
rearch_click_df.head()
```

### .merge()

![join table](14.png)

### รวม 2 ตารางเข้าด้วยกัน

```python
merged_table = rearch_click_df.merge(small_table, how = 'left', left_on = 'business_id', right_on = 'id')
## รวมตารางกันhow ให้ตารางด้านซ้ายเป็นตารางหลัก
## คีย์ของตารางด้านซ้าย คือ business_id
## คีย์ของตารางด้านขวา คือ id
```

```python
merged_table
```
