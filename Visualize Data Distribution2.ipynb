```python
import pandas 
```

```python
irisdata = pandas.read_csv('C:\\Users\\com01\\Documents\\Codepython\\iris.data',header=None) 
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
versicolor_np = irisdata.iloc[50:100,:-1].to_numpy()
virginia_np = irisdata.iloc[100:,:-1].to_numpy()
```

```python
from matplotlib import pyplot as plt
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize'] = [12,12]
plt.subplot(1,4,1) 
plt.boxplot(setosa_np[:,0])
plt.subplot(1,4,2) 
plt.boxplot(setosa_np[:,1])
plt.subplot(1,4,3) 
plt.boxplot(setosa_np[:,2])
plt.subplot(1,4,4) 
plt.boxplot(setosa_np[:,3])
plt.show()
```

## Map

```python
import pandas
```

```python
debt = 'Average Debt.xls.xls'
```

```python
debt_df = pandas.read_excel(debt) #อ่านไฟล์ excel
```

```python
debt_df = pandas.read_excel(debt,skiprows =3) # อ่านไฟล์ excel โดยข้ามไปทั้งหมด 3 แถว
```

```python
debt_df
```

```python
debt_df_debt = debt_df[['Region and province',2549]] # เลือกเอาเฉพาะคอลัมภ์ที่เราสนใจ คือ 'Region and province',2549
```

```python
debt_df_debt.head()
```

```python
debt_df_debt.drop([0,1,2,3,8,31,49,70,85,86,87,88],axis=0,inplace=True) 
# ลบแถวที่ไม่สนใจทิ้งไปเลย axis=0 คือ ดูแนวนอน(แถว) inpleace=True คือ ลบไปเลยไม่ต้องเอาไปใส่ในตัวแปรอื่น
```

```python
debt_df_debt.head()
```

```python
import folium #เป็นคำสั่งที่ใช้แมปแผนที่
import pandas
```

```python
data_to_plot = debt_df_debt # เตรียมข้อมูลไว้ใช้แมปแผนที่
data_to_plot.head()
```

```python
country_geo = 'thai.json' # เอาข้อมูลแผนที่ที่ดาวน์โหลดมาจากเว็บ เข้าไปในโฟลเดอร์ที่เราสร้าง
```

```python
new_col = list()
for d in data_to_plot['Region and province']:
    new_col.append(d.lstrip())
data_to_plot['Region and province']=new_col # ทำการลบช่องว่างข้างหน้าชื่อออกเพื่อที่จะสามารถทำให้ไฟล์ 2 ไฟล์นั้นสามารถเชื่อมกันได้
```

```python
map = folium.Map(location=[10,100], zoom_start=5) # ทำการกำหนดพิกัด ลองติจูดและละติจูดของประเทศไทย
```

```python
map
```

```python
map.choropleth(geo_data=country_geo,data=data_to_plot,
              columns=['Region and province',2549],
              key_on='feature.properties.name',
              fill_color='YlGnBu',fill_opacity=0.7,line_opacity=0.2,
              legend_name='Average Debt')
# ทำการลงสีของข้อมูลที่มีค่าของจำนวนหนี้สินเฉลี่ยต่อครัวเรือนถ้าจังหวัดไหนที่มีจำนวนหนี้สินเฉลี่ยต่อครัวเรือน
#มีค่ามากจะเป็นสีม่วงเข้มส่วนจังหวัดไหนที่มีจำนวนหนี้สินเฉลี่ยต่อครัวเรือนน้อยจะเป็นสีเหลือง
```

```python
map.save('plot_data.html') # ทำการบันทึกแผนที่เป็นเว็บไซต์
```
