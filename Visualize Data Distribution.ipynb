```python
print(f'max = {max(setosa_np[:,3])} , min = {min(setosa_np[:,3])}')
```

```python
colors = [] # สร้าง list มาเก็บค่าที่เป็นไปตามนี้
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b') #ถ้ามีค่าน้อยกว่า 0.15 เป็นสีฟ้า
    elif i < 0.3:
        colors.append('g') #ถ้ามีค่าน้อยกว่า 0.3 เป็นสีเขียว
    elif i <0.45:
        colors.append('y') #ถ้ามีค่าน้อยกว่า 0.45 เป็นสีเหลือง
    else:
        colors.append('r') #ถ้ามีค่านอกเหนือจากนี้ให้เป็นสีแดง
colors
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2] , c= colors) #ให้มีสีตาม list ของสีที่เราสร้างขึ้น
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2] , c= setosa_np[:,3]) # ให้มีสีตามค่าที่ตั้งอัตโนมัติ
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2] , c= setosa_np[:,3] ,cmap = 'YlOrBr')
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2] , c=np.array([[1,0.5,0.8]])) 
#c=np.array([[1,0.5,0.8]]) ใส่สีให้สีผสมกันเอง สีแดงใส่เต็ม = 1 สีเขียวใส่ไป = 0.5 สีฟ้าใส่ไป 0.8 เรียงแบบนี้ตลอด r g b 
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2] , c= setosa_np[:,3] ,cmap = 'YlOrBr', marker = 'p') # marker เบลี่ยนเป็นรูปห้าเหลียม
```

## visualize 2 data เพื่อดูว่าข้อมูล 2 อัน เหมือนหรือแตกต่างกันยังไง

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2] , c= setosa_np[:,3] ,cmap = 'PuBu', marker = 'p')
plt.scatter(versicolor_np[:,0],versicolor_np[:,1],s=30*versicolor_np[:,2] , c= versicolor_np[:,3] ,cmap = 'PuBu', marker = '*')
```

## visualize 3 data เพื่อดูว่าข้อมูล 3 อัน เหมือนหรือแตกต่างกันยังไง
